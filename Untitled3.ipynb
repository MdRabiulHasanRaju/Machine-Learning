```python
import io
import pandas as pd
```

```python
df = pd.read_csv(io.BytesIO(upload['Iris_Data.csv']))
df
```

```python
from  sklearn import preprocessing
labelEncoder = preprocessing.LabelEncoder()
df_labelEncoder = labelEncoder.fit_transform(df['species'])
df_labelEncoder
```

```python
df['species']= df_labelEncoder
df
```

```python
df.isnull().sum()
```

```python
input = df.drop(columns='species')
output = df['species']
```

```python
from sklearn.model_selection import train_test_split
input_train, input_test, output_train, output_test = train_test_split(input, output, train_size = 0.6, test_size = 0.4, random_state = 100)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
```

```python
DT.fit(input_train,output_train)
```

```python
DT.score(input_test, output_test)
```
