```python
import io 
import math
import pandas as pd 
df= pd.read_csv(io.BytesIO(uploaded['diabetes.csv']))
```

```python
df.isnull().sum()
k = int(math.sqrt(df.shape[0]))
k
```

```python
x=df.drop(['Outcome'], axis=1)

y=df['Outcome'] 

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,train_size=0.80, test_size=0.20, random_state=5)

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import chi2

feature= SelectKBest(score_func=f_classif,k='all')
feature.fit(x,y)
pd.DataFrame(feature.scores_)
```

```python
x=df.drop(['Outcome','BloodPressure','SkinThickness'], axis=1)

y=df['Outcome'] 

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,train_size=0.75, test_size=0.25, random_state=200)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=k)
KNN.fit(x_train, y_train)
```

```python
KNN.score(x_test,y_test)
```

```python
#import confusion matrix
from sklearn.metrics import confusion_matrix

y_predict = KNN.predict(x_test)
confusion_matrix(y_test,y_predict)
#OUTPUT
#TP #FP
#FN #TN
```

```python
#classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```
