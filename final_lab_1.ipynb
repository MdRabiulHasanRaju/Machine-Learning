```python
import io 
import pandas as pd 
df= pd.read_csv(io.BytesIO(uploaded['cardio_train.csv']),sep=';')
```

```python
df
# df.isnull().sum()
```

```python
missing = df.weight.mean()
df.weight = df.weight.fillna(missing)

missing = df.ap_hi.mean()
df.ap_hi = df.ap_hi.fillna(missing)

missing = df.ap_lo.mean()
df.ap_lo = df.ap_lo.fillna(missing)

missing = df.cholesterol.mean()
df.cholesterol = df.cholesterol.fillna(missing)

missing = df.gluc.mean()
df.gluc = df.gluc.fillna(missing)

missing = df.smoke.mean()
df.smoke = df.smoke.fillna(missing)

missing = df.alco.mean()
df.alco = df.alco.fillna(missing)

missing = df.active.mean()
df.active = df.active.fillna(missing)

missing = df.cardio.mean()
df.cardio = df.cardio.fillna(missing)
```

```python
df.isnull().sum()
```

```python
x = df.drop(['id', 'cardio'], axis=1)
y = df['cardio']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,train_size=0.8, test_size=0.2, random_state=9)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier()
```

```python
#Fit The Model
RF.fit(x_train,y_train)
```
