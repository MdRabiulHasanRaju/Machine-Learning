```python
import io
import pandas as pd
df= pd.read_csv(io.BytesIO(upload['credit card taiwan svm algorithm.csv']))
```

```python
df
```

```python
df.isnull().sum()
```

```python
missing = df.AGE.mean()
df.AGE = missing
```

```python
NF = df.drop(['ID'],axis=1)
```

```python
x = NF.drop(['default.payment.next.month'], axis=1)
y = NF['default.payment.next.month']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,train_size=0.8, test_size=0.2, random_state=50)
```

```python
from sklearn.svm import SVC
SVM = SVC(gamma='auto')
```

```python
SVM.fit(x_train,y_train)
SVM.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
dataset_size = len(NF)
print(dataset_size)
K = int(dataset_size/200)
CV = cross_val_score(SVM,x,y,cv=K)
print(CV)
```

```python
average = CV.mean()
print(average)
```
