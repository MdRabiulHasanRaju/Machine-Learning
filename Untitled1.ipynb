```python
import io
import pandas as pd
```

```python
df = pd.read_csv(io.BytesIO(upload['Iris_Data -All_Numerical.csv']))
```

```python
df
```

```python
#prepocessing
df.isnull().sum()
```

```python
#specipy input and output
x = df.drop(columns='species') # input = df.drop('species',axis=1)
y = df['species']
print("Value of Input: \n", x.head())
print("Value of Output: \n",y.head())
```

```python
#specify train test variable
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state = 50)
```

```python
#Import Native Bayes From Scikit Learn
from sklearn.naive_bayes import GaussianNB
NB = GaussianNB()
```

```python
#Fit The Model
NB.fit(x_train,y_train)
```

```python
#Evaluate The Model
NB.score(x_test, y_test)
```
