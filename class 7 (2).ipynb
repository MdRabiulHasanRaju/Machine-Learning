```python
import io
import pandas as pd
df= pd.read_csv(io.BytesIO(upload['credit card taiwan svm algorithm.csv']))
```

```python
df
```

```python
df.isnull().sum()
```

```python
missing = df.AGE.mean()
df.AGE = missing
```

```python
NF = df.drop(['ID'],axis=1)
```

```python
x = NF.drop(['default.payment.next.month'], axis=1)
y = NF['default.payment.next.month']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(
    x,y,train_size=0.8, test_size=0.2, random_state=50)
```

```python
from sklearn.svm import SVC
SVM = SVC(gamma='auto')
```

```python
SVM.fit(x_train,y_train)
SVM.score(x_test, y_test)
```

```python
#import confusion matrix
from sklearn.metrics import confusion_matrix

y_predict = SVM.predict(x_test)
confusion_matrix(y_test,y_predict)
```

```python
#classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```
