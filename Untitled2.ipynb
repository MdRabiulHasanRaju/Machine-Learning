```python
import io
import pandas as pd
```

```python
df = pd.read_csv(io.BytesIO(upload['carprices.csv']))
```

```python
df
```

```python
#Import Label Encoder
from  sklearn import preprocessing
#Creating Object
labelEncoder = preprocessing.LabelEncoder()
#Encode
df_labelEncoder = labelEncoder.fit_transform(df['Car Model'])
```

```python
df_labelEncoder
```

```python
#One Hot Encoding
df_OneHotEncoding = pd.get_dummies(df['Car Model'])
df_OneHotEncoding
```

```python
CM = df.drop(columns='Car Model')
```

```python
newData = pd.concat([df,df_OneHotEncoding], axis=1)
```

```python
newData
```

```python
new_df = newData.drop(columns='Car Model')
```

```python
new_df
```

```python
x = new_df.drop(columns='Sell Price')
y = new_df['Sell Price']
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state = 50)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(x_train,y_train)
```

```python
DT.score(x_test, y_test)
```
